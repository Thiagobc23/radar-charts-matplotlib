```python
def plot_img(idx):
    # get URL from df
    url = df.img_url[idx]
    
    # read and plot img
    image = io.imread(url)
    plt.imshow(image)
    
    # plot title n remove axis
    plt.title(df.name[idx], loc='left', fontsize=18)
    ax.axis('off')
```

```python
def plot_radar(idx):
    # categories
    categories = df.iloc[idx,2:].index.tolist()
    N = len(categories) # get number of categories
    
    # values
    values= df.iloc[idx,2:].values.tolist()
    values += values[:1] # repeat first value to close poly

    # calculate angle for each category
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1] # repeat first angle to close poly

    # plot
    plt.polar(angles, values, marker='.') # lines
    plt.fill(angles, values, alpha=0.3) # area
    
    # xticks
    plt.xticks(angles[:-1], categories)

    # yticks
    ax.set_rlabel_position(0) # yticks position
    plt.yticks([5, 10, 15, 20, 25], color="grey", size=10)
    plt.ylim(0,30)
```

```python
# https://www.kaggle.com/sudalairajkumar/daily-temperature-of-major-cities
city = pd.read_csv('data/city_temperature.csv', 
                    usecols=['City', 'Month', 'Year', 'AvgTemperature'])

city_name = 'New York City'
# clean/prepare df
idx = city[city['AvgTemperature'] == -99].index
city.drop(idx, inplace=True)

city = city[city['City'] == city_name]
city = city.groupby(['Year','Month']).mean()
city = city.unstack()

# define fig n ax
fig = plt.figure(figsize=(6,6))
ax = plt.subplot(polar="True")

# define categories
N = 12 # number of categories
categories = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
             'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# loop every for the number of years in the df
for idx in range(len(city)):
    # get values from df and calculate angles
    values = city['AvgTemperature'].iloc[idx].values.tolist()
    angles = [n / float(N) * 2 * pi for n in range(N)]
    # try adding next year, first month value and angle
    try:
        values = values + [city['AvgTemperature'].iloc[idx+1].values[0]]
        angles += angles[:1]
    except:
        continue

    # plot radar
    plt.polar(angles, values, alpha=0.5, lw=0.5, color='black')

    # xticks - needs to be inside the loop to avoid issues with incomplete years.
    plt.xticks(angles[:-1], categories)
    
    
# yticks
ax.set_rlabel_position(0)
plt.yticks([25, 50, 75], color="grey", size=10)
plt.ylim(0,100)

title = '{}\nAvg Temperature - {} to {}'.format(city_name, city.index[0], city.index[-1])
plt.title(title)
plt.show()
```
